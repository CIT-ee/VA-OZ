```python
import pandas as pd

from geopy.distance import vincenty

import datadotworld as dw

import numpy as np
```

```python
ds = dw.load_dataset('batten/va-oz', auto_update=True)

geo_df = ds.dataframes['vacensustracts_geodata_censusbureau']

tract_geoid_list = list(geo_df['geoid'])

all_lats = list(geo_df['intptlat'])

all_lons = list(geo_df['intptlong'])

latlons = zip(all_lats, all_lons)

tract_latlons_list = []

for l in latlons:
    tract_latlons_list.append(l)
```

```python
ds.dataframes['r1r2_universities']
```

```python
def get_latlons_list(file_name):
    df = ds.dataframes[file_name]
    if 'latitude' in df:
        lats_list = list(df['latitude'])
    else:
        lats_list = list(df['y'])
    if 'longitude' in df:
        lons_list = list(df['longitude'])
    else:
        lons_list = list(df['x'])
    latlons_list = list(zip(lats_list, lons_list))
    return latlons_list
```

```python
def get_mean_distance(origins_list, target_file_name, tract_geoid_list):
    i = -1
    distance_dict = {}
    target_latlons_list = get_latlons_list(target_file_name)
    len_target = len(target_latlons_list)
    for origin in origins_list:
        i += 1
        distance_dict[tract_geoid_list[i]] = 0
        sum_from_origin = 0
        for target in target_latlons_list:
            sum_from_origin += vincenty(origin, target).miles
        distance_dict[tract_geoid_list[i]] = sum_from_origin/len_target
    df = pd.DataFrame(list(distance_dict.items()), columns=['geoid', target_file_name[:15] + '_mean_distance'])
    df.to_csv(target_file_name[:15] + '_mean_distance.csv')
    return distance_dict
```

```python
result = get_mean_distance(tract_latlons_list, 'r1r2_universities', tract_geoid_list)
```
